```python
X, y = base_credito_limpa.drop(columns=['id', 'inadimplente']), base_credito_limpa['inadimplente']
```

```python
X.columns
```

```python
cat_ordinal = list(X[['educacao', 'sts_1',
       'sts_2', 'sts_3', 'sts_4', 'sts_5', 'sts_6']])
cat_one = list(X[['sexo', 'est_civil']])
atributos_numericos = list(X.drop(columns=['sexo', 'educacao', 'est_civil', 'sts_1',
       'sts_2', 'sts_3', 'sts_4', 'sts_5', 'sts_6'], axis=1))
feat = [ 'lim_credito', 'sexo', 'educacao', 'est_civil', 'idade', 
                        'sts_1', 'sts_2', 'sts_3', 'sts_4', 'sts_5', 'sts_6',
                        'fatura_1', 'fatura_2', 'fatura_3', 'fatura_4', 'fatura_5', 'fatura_6',
                        'v_pg_1', 'v_pg_2', 'v_pg_3', 'v_pg_4', 'v_pg_5', 'v_pg_6', 'aux']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=42)
```

```python

s_scaler = StandardScaler()
one_encoder = OneHotEncoder(sparse_output=False)
ordinal_encoder = OrdinalEncoder()
```

```python
prepocessor = ColumnTransformer(
    transformers=[
    ('num', s_scaler, atributos_numericos),
    ('one', one_encoder, cat_one),
    ('ord', ordinal_encoder, cat_ordinal)
    ]
)
```

```python
x_prep = prepocessor.fit_transform(X_train)
```

```python
x_test_prep = prepocessor.fit_transform(X_test)
```

```python
model = LogisticRegression(max_iter=1000)
```

```python
model.fit(x_prep, y_train)
```

```python
y_pred = model.predict(x_test_prep)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred_lr_no_pre = model.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred_lr_no_pre))
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_pred_lr_no_pre)
```

```python
dt_classifier = DecisionTreeClassifier(class_weight='balanced', criterion='entropy',
                       max_depth=6, max_leaf_nodes=10)
```

```python
dt_classifier.fit(x_prep, y_train)
y_dt_pred = dt_classifier.predict(x_test_prep)
```

```python
print(classification_report(y_test, y_dt_pred ))
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_dt_pred)
```

```python
params_grid_dt = {'max_depth': np.arange(3, 10),
             'criterion' : ['gini','entropy'],
             'max_leaf_nodes': [5,10,20,100],
             'min_samples_split': [2, 5, 10, 20],
             'class_weight' : ['balanced']}

splitter = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

grid_dt = GridSearchCV(DecisionTreeClassifier(),
                       param_grid=params_grid_dt,
                       scoring="f1_weighted",
                       cv=splitter,
                       verbose=10)

grid_dt.fit(x_prep, y_train)
```
